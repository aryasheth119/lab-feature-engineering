Check for null values in the numerical columns.

```python
numerical = data.select_dtypes(np.number)
numerical.head()
```

Checking for null values

Checking for specific columns which have null values more than zero

```python
# Get the number of null values in each column
null_counts = numerical.isnull().sum()
```

```python
# Select only the columns with null values greater than 0
result_df = numerical[columns_with_null]
```

Cleaning GEOCODE2 column

Handling missing values

```python
data['GEOCODE2'].value_counts()
```

```python
data['GEOCODE2'].fillna('Unknown', inplace=True)
```

```python
print(data['GEOCODE2'].unique())
```

Cleaning WEALTH1 column

```python
data['WEALTH1'].value_counts()
```

```python
# Handling missing values
data['WEALTH1'].fillna(data['WEALTH1'].median(), inplace=True)
```

```python
# Converting WEALTH1 column to numerical data type
data['WEALTH1'] = pd.to_numeric(data['WEALTH1'], errors='coerce')
```

```python
print(data['WEALTH1'].unique())
```

Checking if ADI is numeric

```python
is_numeric = pd.api.types.is_numeric_dtype(data['ADI'])
print("Is ADI numerical?", is_numeric)
```

Cleaning ADI column 

```python
data['ADI'].value_counts()
```

```python
# Handling missing values
specific_value = 0  
data['ADI'].fillna(specific_value, inplace=True)
```

```python
print(data['ADI'].unique())
```

Cleaning DMA and MSA columns

```python
data['DMA'].value_counts()
```

```python
data['MSA'].value_counts()
```
